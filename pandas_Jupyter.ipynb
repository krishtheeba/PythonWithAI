```python
# 1. Load emp.csv to pandas -> df
df_csv = pd.read_csv("D://emp.csv")

# 2. Get info about loaded DataFrame
df_csv.info()   # df_csv.describe()

# 3. List out all row index, column index
print("Row Index : ", df_csv.index)
print("List of column : ", df_csv.columns)

```

```python
# 4. display df shape
print(df_csv.shape)   # tuple

# 5. Test- any duplicates values exists--> drop duplicated update to input DataFrame
df_csv.duplicated()
df_csv.duplicated().sum()
df_csv.drop_duplicates(inplace=True)

# 6. Test - any NaN is exist -> drop index based and column based
print(df_csv.isnull().sum())

# 7. Test - any NaN is exists -> replace row/ column default 
df_csv.dropna(inplace=True)   # default - row based
df_csv.dropna(axis=1) # column based
print(df_csv.isnull().sum())

df_csv.fillna('TestData')   # replace NaN - TestData
df_csv['edept'].fillna('commonDept')
```

```python

# 8. Rename slective columns ( ex: edept==> Dept)
print(df_csv.columns)
df_csv.rename(columns={'edept':'DEPT', 'ecost':'AMOUNT'},inplace=True)
print(df_csv.columns)
```

```python
# 9. Drop selective columns
df_csv.drop(columns=['eid','ecity'], inplace=True)
```

```python
## ELT
#1. Extract product.csv
#2. Load same product.csv --> DB
#3. Transform - use SQL Query
```

```python
# Extract
df_prod= pd.read_csv("D://products.csv")
df_prod.info()
print(df_prod)
```

```python
#Load - Load raw data
import sqlite3
conn= sqlite3.connect("products1.db")
df_prod.to_sql('product_raw',conn,if_exists="replace", index=False)   # Loaded to DB
```

```python
#Transform (inside db) - SQl Query  Not Python programming logic
conn.execute(""" create table product_final as
                 SELECT pid,
                 pname,
                 pcost,
                 CASE
                     when pname= 'pC' THEN pcost * 0.18
                 ELSE pcost
                END AS updatecost
                FROM product_raw;
                """
            )
conn.commit()
```

```python
#Testing
for var in conn.execute("select * from product_final"):
    print(var)
```
