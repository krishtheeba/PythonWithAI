```python
[at5] -> anywhere 'a' (or) char 't' (or) char '5' ->Matched 

^[at5] -> line starts with a (or) t (or) 5

[at5]$ -> line ends with a (or) t (or) 5

[^at5] -> Not matching char a ; Not matching t ; Not matching char 5

[^a-z] ->Not matchin any lowercase chars

[^A-Z] ->Not matching any uppercase chars

[^a-zA-Z] ->Not matching alpha = match space,digits and specialchars

[^a-zA-Z0-9] -> match space and special


[^a-zA-Z0-9\s] -> matching specialchars 

[a-zA-Z0-9] ==>\w

[^a-zA-Z0-9\s] ==>[^\w] - matching specialchars

<Pattern>{n} - n times

ab{2}c => abbc 
ab{3}c ==> abbbc

ab{2,}c => abbc abbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbc //matched 

ab+c ---same as -- ab{1,}c 

ab{2,4}c ==> abbc abbbc abbbbc //matched 
abc
abbbbbc //not-matched 
```
