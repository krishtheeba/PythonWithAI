```python
#Generator
#----------
function - returns an address - generator
           ------------------
                |------yield
```

```python
def fa():
    return 10

def fb():
    yield 10
    yield 20
    

print(type(fa),type(fb()))
```

```python
fa()
```

```python
fb()
```

```python
genObj=fb()
```

```python
print(next(genObj))
print(next(genObj))
print(next(genObj))
```

```python
def fb():
    yield 10
    yield  "D1","D2"
    yield  "D1", ["F1","F2"],["F3","F4"]
    yield  10+20*30
```

```python
genObj=fb()
```

```python
print(next(genObj))
print(next(genObj))
print(next(genObj))
print(next(genObj))
print(next(genObj))
```

```python
genObj= fb()
for var in genObj:
    print(var)
```

```python
genObj=fb()
print(list(genObj))
```

```python
map,filter
|
map(function,collection)
      |         |
     lambda     comprehension/ iterable iter(collection)/ open(inputfile)
```

```python
L=[]
def f1(a):
    return a+100

for var in range(1,6):
    L.append(f1(var))
L
```

```python
map(lambda a:a+100, range(1,6))  # Python 3.x  map()--> generator
                                 # Python 2.x  map()-> []
```

```python
list(map(lambda a:a+100, range(1,6)))
```

```python
d={}
d['K1']=list(map(lambda a:a+100, range(1,6)))

L=[]
L.append(list(map(lambda a:a+100, range(1,6))))

print(L)
print(d)
```

```python
map(function,collection) -> generator
|
filter(function,collection) -> generator
-------// return True  value only
```

```python
list(map(lambda a: a> 15,  [10,20,15,9,2,3]))
```

```python
list(filter(lambda a: a> 15,  [10,20,15,9,20,2,3]))
```
