```python
Task
====
Write a program
Step 1: Initialize pin number ( ex: pin=1234)
Step 2: Use while loop to iterate following statements thrice
-------> (i) Read the Input pin
-------> (ii) Compare input pin with existing pin
-------> (iii) If both pin match , display - pin number is matched at count  & exit from loop(break)
-------> (iV) If all attempts doesnot match, display - Your pin is blocked
```

```python
pin=1234
count=0
while(count < 3):
    count = count + 1    # count+=1   compound assignment
    p=int(input("Enter the Pin:"))
    if(p == pin):
        print(f"Success - pin is matched at -{count}")
        break

if(p != pin):
    print("Sorry Your pin is Blocked")
```

```python
Task
====
Modify the Pin example
-> create a new file - pin_history.log -  append mode
-> write user input pin(success/fail/block) - count - date/time 
open pin_history.log - see input details
```

```python
import time
wobj=open('pin_history.log', 'a')

pin=1234
count=0

while(count < 3):
    count = count + 1    # count+=1   compound assignment
    p=int(input("Enter the Pin:"))
    if(p == pin):
        wobj.write(f'Success - pin {p} is matched at {count} date/time :{time.ctime()}\n')
        print(f"Success - pin is matched at -{count}")
        break
    else:
        wobj.write(f'Failed-user input pin {p} Not Matched - {count} date/time:{ time.ctime()}\n')

if(p != pin):
    wobj.write(f'Sorry your pin is blocked- {count} date/time :{time.ctime()}\n')
    print("Sorry Your pin is Blocked")
wobj.close()
```
