```python
import time
class model_provider:
    '''Model Attributes'''
    model_name=''
    model_vector_size=''
    model_size=''
    model_download_count=0
    def __init__(self,model_name,model_vector_size,model_size,model_download_count):
        self.model_name=model_name
        self.model_vector_size=model_vector_size
        self.model_size=model_size
        self.model_download_count=model_download_count
        print("Initialization Done")

    def display(self):
        '''display model details'''
        print(self.model_name,self.model_vector_size, self.model_size, self.model_download_count) 

    def __call__(self):  # object as callable obj
        return self.model_name,self.model_vector_size, self.model_size, self.model_download_count  # as tuple - python default multiplevalue

    def update_vector_size(self,model_vector_size):
        self.model_vector_size=model_vector_size
        s1=self.model_name +"," +self.model_vector_size
        s2=time.ctime()

        with open("D://model_history.log","a") as wobj:
            wobj.write(s1+s2+"\n")

obj1= model_provider("Google",'5T','10GB',100)
obj1()   # this invoke __call__()
#obj1.display()
#obj1.update_vector_size('10T')
#obj1.display()
print(type(obj1()) )      
        
```

```python
Decorator- add functionality to existing classes/ function
        - function that wraps other function / classes
```

```python
class C:
    def func():
        """No self here"""
        print("Method used as function")
    func=staticmethod(func)

c=C()
c.func()
```

```python
class C:
    @staticmethod
    def func():
        """No self here"""
        print("Method used as function")
    
    @staticmethod
    def func2():
        """No self here"""
        print("Method used as function")
c=C()
c.func()
c.func2()
```

```python
def f1():   # decorator
    def fx():   # wrapper fn
        def appl1():
            print("Application 1")
        def appl2():
            print("Application 2")
        appl1()
        appl2()
    return fx

fx=f1()
fx()    
print(f1)
```

```python
def news(a):   # decorator
    def fx():   # wrapper fn
        def City1():
            print("Weather City 1")
        def City2():
            print("Weather City 2")
        City1()
        City2()
        a()   # City3()
    return fx
```

```python
@classmethod  - method accepts cls as first arg
```

```python
class Eg:
    @classmethod
    def show(cls):
        print("This is class method of",cls)
Eg.show()
```

```python
class Eg:
   def show(self):  # self here refers Obj
        print("This is class method of",self)
obj=Eg()
obj.show()
```

```python
class Eg:
    @staticmethod
    def show():
        print("This is static method")

Eg.show()
```

```python
Instance Method  --  self (Object) -> call by   obj.method()
Class Method    --   cls ( Class)  -> Class.method() (or)  Obj.method()   @classmethod
Static Method   --   None    -> Class.method()       @staticmethod
```
