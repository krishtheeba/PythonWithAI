```python
class A:
    def f1(self):
        print("A->f1")
class B(A):
    def f2(self):
        print("B->f2")

childObj=B()
childObj.f1() 
childObj.f2()
      
```

```python
class A:
    def f1(self):
        print("A->f1")
class B(A):
    def f1(self):
        print("B->f1")  # Method Overidding- oop- Runtime Polymorphism

childObj=B()
childObj.f1() # child f1
```

```python
L=[A(),B(),B(),A()]
for var in L:
    var.f1()
```

```python
class A:
    def f1(self):
        print("A->f1")
class B(A):
    def f1(self):
        super().f1() # explicitly call to parent
        

childObj=B()
childObj.f1() # child f1-->parent f1
```

```python
class Person:
    def __init__(self,name):
        self.name=name
        print("Person Constructor called")

class Student(Person):
    def __init__(self,name,grade):
        super().__init__(name) # first line
        self.grade=grade
        print("Child constructor Done")

childObj=Student("Arun","A")
```

```python
class Person:
    
    def __init__(self,name):
        self.name=name
        print("Person Constructor called")
    def f1(self):
        print(self.name)

class Student(Person):
    def __init__(self,name,grade):
       # super().__init__(name) # first line
        self.grade=grade
        print("Child constructor Done")

childObj=Student("Arun","A")
childObj.f1()
```
